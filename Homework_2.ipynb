```
#Problem 1 and 2

import numpy as np

x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print("Array x + y:",x + y)
print("Array 2*x:",2*x)
print("Array x*y:",x*y)
print("Array x dot y:",np.dot(x,y))
```

```
#Problem 3 and 4

M =  [(1, 2, 3), (2, 1, 3), (3, 2, 1)]
N =  [(4, 5, 6), (5, 4, 6), (6, 5, 4)]
MN = np.matmul(M,N)
print("Matrix M * Matrix N:\n",MN)
print("Matrix N * Matrix M:\n",np.matmul(N,M))
M2 = np.linalg.inv(M)
N2 = np.linalg.inv(N)
print("Inverse of Matrix M:\n",M2)
print("Inverse of Matrix N:\n",N2)
print("Inverse of Matrix N * Matrix M:\n",np.matmul(M2,M))
print("Inverse of Matrix N * Matrix N:\n",np.matmul(N2,N))
# Printed out the matrixes a little odd but still is the identiy matrix
print("Matrix M * N * transpose of array x:\n", np.matmul(MN,np.transpose(x)))
```

```
#Problem 5 and 6

import pandas as pd 
import numpy as np
from sklearn.datasets import fetch_california_housing
housing  = fetch_california_housing()
housing_pd = pd.DataFrame(data= np.c_[housing['data'], housing['target']], columns= housing['feature_names'] + ['housing_value'])
housing_pd.head()
print("MedInc mean:",housing_pd["MedInc"].mean())
print("Housing value mean:",housing_pd["housing_value"].mean())
print("MedInc median:",housing_pd["MedInc"].median())
print("Housing value median:",housing_pd["housing_value"].median())
print("Housing value quantiles:", housing_pd["housing_value"].quantile([0.25,0.5,0.75]))
print("MedInc quantiles:",housing_pd["MedInc"].quantile([0.25,0.5,0.75]))
print("Info:")
print(housing_pd.info())
print("Describe: ")
print(housing_pd.describe())
# mean and medians from before match with the table from describe
```
